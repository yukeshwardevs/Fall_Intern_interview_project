```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

def remove_integers_and_brackets(text):
    # I'm preprocessing data to extract only text
    return ''.join(char for char in text if not char.isdigit() and char not in ['[', ']'])

def process_csv(input_file, output_file):
    with open(input_file, 'r') as f:
        reader = csv.reader(f)
        with open(output_file, 'w') as out_f:
            writer = csv.writer(out_f)
            for row in reader:
                for cell in row:
                    modified_cell = remove_integers_and_brackets(cell.strip())
                    if modified_cell:
                        writer.writerow([modified_cell])

input_file = '/content/drive/MyDrive/Bollywood-Data-master/wikipedia-data/male_adjverb.csv'
output_file = '/content/drive/MyDrive/Bollywood-Data-master/wikipedia-data/outputuh.csv'

process_csv(input_file, output_file)
df = pd.read_csv(output_file, header=None)

# I'm flattening the DataFrame to count occurrences
text_counts = df[0].value_counts().head(30)

plt.figure(figsize=(10, 6))
text_counts.plot(kind='bar')
plt.title('Male adjverb (top 30)')
plt.xlabel('Text')
plt.ylabel('Occurrences')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/Bollywood-Data-master/wikipedia-data/songsDB.csv", usecols=[2])

#I'm preprocessing the given data
data['gender'] = data['gender'].str.lower()

# I'm replacing other gender values with 'other'
data.loc[~data['gender'].isin(['male', 'female']), 'gender'] = 'other'

gender_counts = data['gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution in Songs in the movie\n')
plt.axis('equal')
plt.show()
```
